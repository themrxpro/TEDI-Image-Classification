```python
import tensorflow as tf
import pickle

model = tf.keras.models.load_model('./model_2_0')

history = pickle.load(open('./model_2_0_trainHistory', "rb"))
```

```python
print(history.keys())
print(len(history['accuracy']))
```

```python
from matplotlib import pyplot as plt

plt.title('TITLE')
plt.plot(   range(1,21),    history['val_accuracy'],
            range(1,21),    history['accuracy']
            )

plt.legend(['val_accuracy','accuracy'])
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.show
```

```python
plt.title('TITLE')
plt.plot(   range(1,21),    history['val_loss'],
            range(1,21),    history['loss']
            )
            
plt.legend(['val_loss','loss'])
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.show
```
